```python
# Remove duplicate rows from the DataFrame in place
df.drop_duplicates(inplace=True)
```

```python
# Drop the 'house_rules' and 'license' columns from the DataFrame in place
df.drop(['house_rules','license'],axis=1,inplace=True)
```

```python
df.rename(columns={
    'price':'price(USD)',
    'service fee':'service fee(USD)'
}, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['price(USD)']=df['price(USD)'].astype(float)
df['service fee(USD)']=df['service fee(USD)'].astype(float)
df['id']=df['id'].astype(int)
df['host id']=df['host id'].astype(str)
df['last review']=pd.to_datetime(df['last review'])
df['Construction year']=df['Construction year'].astype(int)
```

```python
df.loc[df['neighbourhood group']=='brookln','neighbourhood group']='Brooklyn'
```

```python
df=df.drop(df[df['availability 365']>500].index)
```

```python
# Get the unique values from the 'room type' column
property_types = df['room type'].unique()

# Print the unique property types
print("Different property types in the dataset:")
for prop_type in property_types:
    print(prop_type)
```

```python
# Count the occurrences of each neighbourhood group
neighbourhood_counts = df['neighbourhood group'].value_counts()

# Find the neighbourhood group with the highest count
highest_listings_neighbourhood = neighbourhood_counts.idxmax()
highest_listings_count = neighbourhood_counts.max()

# Print the result
print(f"The neighbourhood group with the highest number of listings is '{highest_listings_neighbourhood}' with {highest_listings_count} listings.")
```

```python
# Count the occurrences of each room type
room_type_counts = df['room type'].value_counts().reset_index()
room_type_counts.columns = ['room type', 'count']

# Create a bar plot of room type counts
plt.figure(figsize=(10, 6))
sns.barplot(x='room type', y='count', data=room_type_counts)
plt.title('Count of Each Room Type')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot of neighbourhood group counts
plt.figure(figsize=(10, 6))
sns.barplot(x=neighbourhood_counts.index, y=neighbourhood_counts.values)
plt.title('Count of Listings per Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate the average price for each neighbourhood group
average_price_by_neighbourhood = df.groupby('neighbourhood group')['price(USD)'].mean()

# Find the neighbourhood group with the highest average price
highest_avg_price_neighbourhood = average_price_by_neighbourhood.idxmax()
highest_avg_price = average_price_by_neighbourhood.max()

# Print the result
print(f"The neighbourhood group with the highest average price is '{highest_avg_price_neighbourhood}' with an average price of ${highest_avg_price:.2f}.")

# Optional: Display the average prices for all neighbourhood groups
# display(average_price_by_neighbourhood.sort_values(ascending=False))
```

```python
# Create a bar plot of average price per neighbourhood group
plt.figure(figsize=(7, 6))
sns.barplot(x=average_price_by_neighbourhood.index, y=average_price_by_neighbourhood.values)
plt.title('Average Price per Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate the average price for each construction year
average_price_by_year = df.groupby('Construction year')['price(USD)'].mean().reset_index()

# Create a line plot of average price per construction year
plt.figure(figsize=(12, 6))
sns.lineplot(x='Construction year', y='price(USD)', data=average_price_by_year)
plt.title('Average Price per Construction Year')
plt.xlabel('Construction Year')
plt.ylabel('Average Price (USD)')
plt.tight_layout()
plt.show()
```

```python
# Calculate the average review rate for verified vs. unverified hosts
average_review_rate_by_verification = df.groupby('host_identity_verified')['review rate number'].mean()

# Display the result in a table
display(average_review_rate_by_verification)
```

```python
# Create a boxplot to compare review rates for verified vs. unverified hosts
plt.figure(figsize=(8, 6))
sns.boxplot(x='host_identity_verified', y='review rate number', data=df)
plt.title('Review Rate Distribution by Host Identity Verification Status')
plt.xlabel('Host Identity Verified')
plt.ylabel('Review Rate Number')
plt.tight_layout()
plt.show()
```

```python
# Create a regression plot to visualize the relationship between price and service fee
plt.figure(figsize=(7, 5))
sns.regplot(x='price(USD)', y='service fee(USD)', data=df, scatter_kws={'alpha':0.3})
plt.title('Relationship between Price and Service Fee')
plt.xlabel('Price (USD)')
plt.ylabel('Service Fee (USD)')
plt.tight_layout()
plt.show()
```

```python
# Calculate the average review rate by neighbourhood group and room type
average_review_rate_by_group_and_type = df.groupby(['neighbourhood group', 'room type'])['review rate number'].mean()

# Display the result
display(average_review_rate_by_group_and_type)
```

```python
# Calculate the overall average review rate
overall_average_review_rate = df['review rate number'].mean()

print(f"Overall average review rate: {overall_average_review_rate:.2f}")

# Calculate the average review rate by neighbourhood group
average_review_rate_by_neighbourhood = df.groupby('neighbourhood group')['review rate number'].mean()

print("\nAverage review rate by neighbourhood group:")
display(average_review_rate_by_neighbourhood)

# Calculate the average review rate by room type
average_review_rate_by_room_type = df.groupby('room type')['review rate number'].mean()

print("\nAverage review rate by room type:")
display(average_review_rate_by_room_type)
```

```python
# Convert the Series to a DataFrame for plotting
average_review_rate_df = average_review_rate_by_group_and_type.reset_index()
average_review_rate_df.columns = ['neighbourhood group', 'room type', 'average review rate']

# Pivot the DataFrame for plotting a grouped bar chart
pivot_table = average_review_rate_df.pivot(index='neighbourhood group', columns='room type', values='average review rate')

# Create a grouped bar chart
ax = pivot_table.plot(kind='bar', figsize=(12, 7))

plt.title('Average Review Rate by Neighbourhood Group and Room Type')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Review Rate')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Room Type')
plt.tight_layout()
plt.show()
```

```python
# Create a regression plot to visualize the relationship between calculated host listings count and availability 365
plt.figure(figsize=(10, 6))
sns.regplot(x='calculated host listings count', y='availability 365', data=df, scatter_kws={'alpha':0.3})
plt.title('Relationship between Calculated Host Listings Count and Availability 365')
plt.xlabel('Calculated Host Listings Count')
plt.ylabel('Availability 365')
plt.tight_layout()
plt.show()
```
